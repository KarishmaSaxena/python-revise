```python
#label encoding
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
file_path=('/content/breast-cancer.csv')
data= pd.read_csv(file_path)
data.head()
```

```python
data['perimeter_mean'].value_counts()
```

```python
data['diagnosis'].value_counts()
```

```python
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(data.diagnosis)
```

```python
#appending labels to dataframe
data['target'] = labels
print(data.head())
```

0-> Benign
1-> Malignant
