Pandas Libraby:

     → data manipulation and data analysis
     → Built on top of NumPy

Key Features of Pandas

      → Handles missing data (NaN)
      → Powerful data filtering & selection
      → Easy data cleaning & preprocessing
      → Supports grouping, aggregation, merging
      → Fast input/output (CSV, Excel, JSON, SQL)

Pandas DataFrame (structured tables)

      → A two-dimensional labeled data structure
      → Consists of rows and columns
      → Each column can have different data types
      → Similar to an Excel sheet / SQL table
      → Built on top of NumPy
