    Central Tendency
      -> Mean -> Sum of all values divided by the number of values.
      -> Median -> Middle value when data is sorted
      -> Mode -> Most frequently occurring value in the dataset



```python
print(spaceship_datset['Spa'].mean())
print(spaceship_datset['Spa'].median())
print(spaceship_datset['Spa'].mode())
```

Imputation -> filling missing values -> mean,meadian,mode

dropping    -> removing rows or columns that contain missing values
            ->Missing values are very high (>40-50%)
            -> Dataset is huge
            -> Feature has no impact on prediction

```python
#replacing missing values with median value
spaceship_datset['Spa'].fillna(spaceship_datset['Spa'].median())
```

```python
spaceship_datset['FoodCourt'].fillna(spaceship_datset['FoodCourt'].median())
```

fillna(...)
 replaces missing values with that mean

inplace=True
 saves the changes directly in the DataFrame

```python
#Filling missing value with mean
import pandas as pd
spaceship_datset=pd.read_csv('/content/test.csv')
print(spaceship_datset.head())
spaceship_datset['FoodCourt'].fillna(spaceship_datset['FoodCourt'].mean(),inplace=True)
```

```python
# filling missing value with mode
import pandas as pd
spaceship_datset['VRDeck'].fillna(spaceship_datset('VRDeck').mode(),inplace=True)
```

```python
#droping All Rows with Any Missing Value
spaceship_datset.dropna(inplace=True)
print(spaceship_datset)
```

```python
#Dropping Columns with Missing Values
spaceship_datset.drop(columns=['Cabin'], inplace=True)
print(spaceship_datset.columns)
```
