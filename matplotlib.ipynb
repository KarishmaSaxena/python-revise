Matplotlib -> for making plots (data visualization)

Why

       -> Data in numbers is hard to understand
       ->Graphs make insights clear & quick

pyplot provides a MATLAB-like interface for plotting.

Common Types of Plots
                  -> Line Plot
                  -> Scatter Plot
                  -> Bar Plot
                  -> Histogram
                  -> Pie Chart


    Matplotlib
            → Python library for data visualization
            → Used to create 2D & basic 3D plots
            → Converts numerical data into graphs
            → Helps in data analysis & pattern detection
            → Works well with NumPy & Pandas
            → Common plots: line, bar, scatter, histogram, pie


x
  → Independent variable
  → Values along x-axis
  x → input

y
  → Dependent variable
  → Computed from x
  y → output

  Every time x changes, y automatically changes

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,10,100)
y= np.sin(x)
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data
Scatter plot - Used to show relationship between two variables

```python
# sin wave
plt.plot(x,y)
plt.show()
# scatter plot
plt.scatter(x, y)
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title to x-axis, y-axis
plt.xlabel("inputs(angle )")
plt.ylabel("outputs(sine values)")
plt.title("sine wave")
plt.plot(x,y)
plt.show()
```

```python
#parabola
x= np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# for showing graph with symbols +
plt.plot(x,y,"r+")
plt.show()
```

```python
#hyperbola
x1 = np.linspace(-10, -0.1, 20)
x2 = np.linspace(0.1, 10, 20)
y1 = 1/x1
y2 = 1/x2
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
# bar plot
languages = ["C", "C++", "Java", "Python", "Javascript"]
popularity = [22.2, 17.6, 8.8, 8, 7.7]
plt.title("Popularity of Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.bar(languages,popularity)
plt.show()
```

autopct ->>>

It tells Matplotlib to automatically print percentages on each slice of the pie chart.

auto ->>
→ automatically
 pct ->>
→ percentage

% → format specifier
f → floating-point number

%1.1f ->>
        1 → minimum width
       .1 → 1 digit after decimal
        f → float

```python
# Pie Chart
age = [22, 55, 62, 45, 21, 22, 34,60]
level = ["infant" , "teen" , "adult" , "adult" , "teen" , "infant" , "old","very-old"]
plt.pie(age,labels=level,autopct="%1.1f%%")
plt.title("Age Group Distribution")
plt.show()
```

```python
#scatter plot
x=np.linspace(10,20,20)
y=np.sin(x)
z=np.cos(x)
plt.title("Scater Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.scatter(x,y,color="red")
plt.scatter(x,z,color="green")
plt.show()
```

    Figure->>

    -> It is the entire window where plots appear
    -> Figure = notebook page

    axes ->>
      -> the actual plot area
      -> Axes is the area where data is drawn (with X, Y, Z axes).
      -> You can have multiple axes in one figure.

    projection ->
    -> tells Matplotlib how to view the plot
    -> Common projections:
                         -> rectilinear → default 2D plot
                         ->  3d → 3D plot (X, Y, Z)

```python
fig = plt.figure()              # create canvas
ax = plt.axes(projection="3d")  # create plot area
ax.scatter(x, y, z)             # draw data
plt.show()                      # display
```
