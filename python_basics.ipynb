```
#functions
number = int(input("enter a factorial number"))
fact=1
number=number+1
if(number==0):
  print("factorial of 0 is 1")
for i in range(fact,number): # fact =1 , number =6+1->7 range(1,7)=> 1,2,3,4,5,6=720
  fact=fact*i
print(fact)
```

```
num = 7

def factorial(num):
  fact=1
  for i in range(fact,num+1):
    fact=fact*i
  return fact
print(factorial(num))
```
