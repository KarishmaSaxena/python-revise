```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# built-in data set
tips= sns.load_dataset("tips")
print(tips)
```

```python
print(tips.head())
```

```python
#default theme for plots
sns.set_theme()
```

```python
sns.relplot(data=tips,x="total_bill",y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load irisdataset
iris_dataset= sns.load_dataset("iris")
print(iris_dataset)
```

```python
sns.relplot(data=iris_dataset,x='sepal_length',y='petal_length',hue='species')
sns.relplot(data=iris_dataset, x='sepal_width',y='petal_width',hue='species')
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
house = fetch_california_housing()
print(house)
housing=pd.DataFrame(house.data,columns=house.feature_names)
housing['Price']=house.target
```

```python
print(housing.head())
```

```python
#distribution plot
import seaborn as sns
sns.displot(housing['Longitude'])
sns.displot(housing['Latitude'])
sns.displot(housing['Price'])
```

```python
#correaltion
housing.corr()
```

    Heatmap
       Visualization technique
       Uses color intensity to represent values
       Helps identify relationships, correlations, and patterns in data

    parameter
    data  Matrix or DataFrame to visualize (e.g., correlation matrix)
    cmap  Color scheme used in the heatmap
    annot  Shows values inside each cell
    fmt  Format of annotated values (e.g., .2f)
    linewidths  Width of lines between cells
    linecolor  Color of grid lines
    cbar  Displays or hides color bar
    cbar_kws  Custom settings for color bar
    square  Makes each cell a perfect square
    vmin  Minimum value for color scale
    vmax  Maximum value for color scale
    center  Midpoint value of color map
    mask  Hides selected cells of the heatmap
    xticklabels  Shows or hides x-axis labels
    yticklabels  Shows or hides y-axis labels
    robust  Ignores extreme values for color scaling
    alpha  Transparency of the heatmap
    ax  Axis on which the heatmap is drawn

```python
#heat map
plt.figure(figsize=(10,10))
sns.heatmap(housing.corr(),cmap="Blues",annot=True,fmt=".2f",linewidths=0.5,square=True)
```
