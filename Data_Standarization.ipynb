    Data standardization
         1 Data standardization is a data preprocessing technique.
         2 It rescales numerical features to a common scale.
         3 After standardization, mean becomes 0.
         4 Standard deviation becomes 1.
         5 It ensures all features contribute equally to the model.
         6 Prevents features with large values from dominating the model.
         7 Commonly used in machine learning algorithms like regression, SVM, KNN, and neural networks.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

dataset= load_breast_cancer()
print(dataset)
#loading data to pandas dataframe
data_frame = pd.DataFrame(dataset.data, columns=dataset.feature_names)
print(data_frame)
data_frame.shape
# here x-> features , y-> target
X=data_frame
Y=dataset.target
print(X)
print(Y)
#Splitting the data before standardizating it into train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X.shape,X_train.shape,X_test.shape)
# data is not standardize
print(dataset.data.std())
#to standardaize
scaler = StandardScaler()
scaler.fit(X_train)
X_train_standardize = scaler.transform(X_train)
print(X_train_standardize)
X_test_standardize = scaler.transform(X_test)
print(X_test_standardize.std())
```
