```python
import matplotlib.pyplot as plt
```

```python
#line plot
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,5,6,8,9,10]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot')
plt.plot(x,y)
plt.show()
```

```python
#bar plot
x=[10,20,30,40,50]
y=["english","hindi","physics","chemistry","bio"]
plt.bar(x,y)
plt.title('bar plot')
plt.xlabel('marks')
plt.ylabel('subject')
plt.show()
```

```python
#Scatter plot
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot ')
plt.show()
```

```python
#Histogram
age=[22,10,1,3,5,33,45,63,86,99,12,14,15,67,88,55,22,34,65]
plt.hist(age,bins=30)
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```

    Boxplot
      => Used to detect outliers and spread
            
    contains
             -> Minimum -> lower whisker) -> 1
             -> First Quartile (Q1) -> Value below which 25% of the data lies , Left/bottom edge of the box
             -> Median (Q2) -> The middle value Shown as a line inside the box Divides data into two equa
             ->Third Quartile (Q3) -> Value below which 75% of the data lies Right/top edge of the box
             -> Maximum (upper whisker) -> 99

```python
#box plot
plt.boxplot(age)
plt.show()
```

```python
#pie chart
labels=["yes","no"]
sizes=[30,70]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```
